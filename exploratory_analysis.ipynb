```python
# %% [markdown]
# # Exploratory Data Analysis - Credit Card Fraud Detection

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# %%
# Load data
df = pd.read_csv('../data/creditcard.csv')
print(f"Dataset Shape: {df.shape}")
df.head()

# %%
# Check Class Imbalance
print(df['Class'].value_counts())
print(f"Fraud Percentage: {df['Class'].mean() * 100:.3f}%")

# Plot Class Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Transaction Class Distribution (0: Legit, 1: Fraud)')
plt.yscale('log') # Log scale because fraud is extremely small
plt.show()

# %%
# Distribution of Transaction Amounts
plt.figure(figsize=(10,4))
sns.histplot(df[df['Class'] == 1]['Amount'], color='red', label='Fraud', kde=True, bins=50)
sns.histplot(df[df['Class'] == 0]['Amount'], color='blue', label='Legit', kde=True, bins=50)
plt.title('Distribution of Transaction Amounts')
plt.xlim(0, 2000)
plt.legend()
plt.show()

# %%
# Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix')
plt.show()
```
